#                                           # Outlier Treatment

Outlier detection & removal is essential for data preprocessing which make the things easy while proceeding further

```python
f, ax = plt.subplots(figsize=(20,8))
sns.boxplot(x=df["Name"].values, y = df["km_driven"].values,palette="CMRmap",ax=ax)
plt.xticks(rotation=90)
plt.show()
```

```python
def cap_data(df):
    for col in df.columns:
        print("capping the ",col)
        if (((df[col].dtype)=='float64') | ((df[col].dtype)=='int64')):
            percentiles = df[col].quantile([0.01,0.99]).values
            df[col][df[col] <= percentiles[0]] = percentiles[0]
            df[col][df[col] >= percentiles[1]] = percentiles[1]
        else:
            df[col]=df[col]
    return df

final_df=cap_data(df)
```

#                                             # Feature Selection

Feature selection is a process of selecting the most important/relevent features from the data which will have a coorelation with the other variable.

```python
df= df.reindex(columns=['Name','year','selling_price','km_driven','fuel','seller_type','transmission','owner'])
```

#                                                # Train Test Split

Before processing further will split the data into train and test so that the model will applicable in train will check the performance on the test set

```python
x=df.drop(columns=['selling_price'],axis=1)
y=df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
y
```

#                                                     # Scaling

scaling is the process of converting data to a common range of values. This is necessary for most machine learning algorithms to function correctly.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

#                                                # Model Building

We will now jump to model building which is on of the most important step and will use RandomForest Regressor.

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train,y_train)
```

```python
pred_rfr = rfr.predict(x_test)
```

#                                              # Model Performance

I have build our model. Now will check how the model is performing with the help of the below coded metrics.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
import sklearn.metrics as metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, pred_rfr))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, pred_rfr))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, pred_rfr)))
print(r2_score(y_test,pred_rfr))
```

The mean absolute error for our algorithm is 121368.67059528963, which is less than 12 percent of the mean of all the values in the selling_price column. This means that our algorithm made a prediction, but it needs a lot of improvement.

```python
df=pd.DataFrame({'Actual':y_test, 'Predicted':pred_rfr})
df
```
